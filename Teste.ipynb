```python
resultado_a = []
for cep,peso in pedidos.iloc[:, [4, 7]].values:
    cep_limpo = int(cep.replace("-","").replace("'",""))
    linha = carries_a[(cep_limpo >= carries_a['ini_CEP']) & (cep_limpo <= carries_a['fim_CEP'])]
    valor_a = linha.loc[:, peso].values
    resultado_a.append(valor_a)
```

```python
pedidos['LEADTIME_A'] = 0
pedidos['LEADTIME_B'] = 0
pedidos.dtypes
```

```python
tempo_entrega_a = []
for cep in pedidos.ZIPCODE.values:
    cep_limpo = int(cep.replace("-","").replace("'",""))
    linha = carries_a[(cep_limpo >= carries_a['ini_CEP']) & (cep_limpo <= carries_a['fim_CEP'])]
    valor_a = linha.lead_time.values
    tempo_entrega_a.append(valor_a)
```

```python
df_resultado = pd.DataFrame(tempo_entrega_a)
pedidos['LEADTIME_A'] = df_resultado
pedidos.head(40)
```
